# 🔍 easy-rag — Quickstart Notebook

Yeh notebook RAG ke 3 steps sikhata hai:
1. **Ingest** — Documents load karo
2. **Retrieve** — Relevant chunks dhundho
3. **Generate** — Answer banao


```python
# Step 0: Setup
import sys; sys.path.insert(0, '../src')
from rag_engine import RAGPipeline

# Apni API key set karo
import os
os.environ['OPENAI_API_KEY'] = 'sk-...'  # Replace with your key

rag = RAGPipeline(provider='openai', store_path='../vector_store_nb.json')
print('✅ RAG pipeline ready!')
```

## Step 1: Documents Ingest karo

```python
# Option A: Text directly paste karo
sample_text = """
Retrieval-Augmented Generation (RAG) ek AI technique hai jisme:
1. Documents ko vector embeddings me convert kiya jata hai.
2. User ke question ke similar chunks retrieve kiye jate hai.
3. Woh chunks LLM ko context ke taur pe diye jate hai.
4. LLM woh context use karke accurate answer deta hai.

RAG hallucinations reduce karta hai aur domain-specific Q&A enable karta hai.
Common use cases: customer support, internal knowledge base, document QA.
"""

rag.ingest_text(sample_text, source='rag-intro')

# Option B: File ingest
# rag.ingest_file('../data/sample_docs/company_faq.txt')

# Option C: PDF
# rag.ingest_pdf('../data/my_report.pdf')

# Option D: Poora folder
# rag.ingest_folder('../data/sample_docs')
```

## Step 2: Retrieve — Relevant chunks dekho

```python
query = 'RAG kya hai aur kaise kaam karta hai?'
chunks = rag.retrieve(query)

print(f'Top {len(chunks)} chunks found:\n')
for i, c in enumerate(chunks, 1):
    print(f'[{i}] Source: {c["metadata"]["source"]}')
    print(c['text'][:300])
    print('---')
```

## Step 3: Generate — Answer lo!

```python
result = rag.ask('RAG ke main use cases kya hain?')

print('🤖 Answer:')
print(result['answer'])
print('\n📚 Sources:', result['sources'])
```

## Bonus: Custom System Prompt

```python
# Apna custom system prompt daal sakte ho
custom_prompt = """
You are a RAG expert assistant. Answer in simple Hindi/English mix (Hinglish).
Always explain technical terms simply. Use bullets where appropriate.
If information is not in context, say 'Yeh information available nahi hai.'
"""

result = rag.ask(
    'RAG aur fine-tuning me kya difference hai?',
    system_prompt=custom_prompt
)
print(result['answer'])
```
